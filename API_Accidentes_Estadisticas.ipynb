```python
"""
At the command line, only need to run once to install the package via pip:

$ pip install google-generativeai
"""

import google.generativeai as genai

my_api_key = "xxxxxxxxxxxxxxxxxxxxx"

genai.configure(api_key=my_api_key)

# Set up the model
generation_config = {
  "temperature": 0.9,
  "top_p": 1,
  "top_k": 1,
  "max_output_tokens": 2048,
}

safety_settings = [
  {
    "category": "HARM_CATEGORY_HARASSMENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_HATE_SPEECH",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_SEXUALLY_EXPLICIT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
  {
    "category": "HARM_CATEGORY_DANGEROUS_CONTENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE"
  },
]

model = genai.GenerativeModel(model_name="gemini-1.0-pro",
                              generation_config=generation_config,
                              safety_settings=safety_settings)
convo = model.start_chat(history=[
  {
    "role": "user",
    "parts": ["cuantos accidentes de trabajo hubo en Argentina en 2023"]
  },
  {
    "role": "model",
    "parts": ["**No tengo acceso a información en tiempo real, por lo que no puedo proporcionar datos sobre el número de accidentes de trabajo en Argentina en 2023.**\n\nPara obtener la información más actualizada, te recomiendo consultar fuentes oficiales, como el Ministerio de Trabajo, Empleo y Seguridad Social de Argentina o el Instituto Nacional de Estadística y Censos (INDEC)."]
  },
  {
    "role": "user",
    "parts": ["cuantos accidentes de trabajo y por categoria hubo en Argentina en el 2021?"]
  },
  {
    "role": "model",
    "parts": ["Según el **Informe Anual de Estadísticas de Accidentes de Trabajo y Enfermedades Profesionales** publicado por la Superintendencia de Riesgos del Trabajo (SRT) de Argentina, en el año 2021 se registraron **622.222 accidentes de trabajo** y **16.340 enfermedades profesionales**.\n\n**Distribución por categoría de los accidentes de trabajo:**\n\n* **Accidentes in itinere:** 22,6%\n* **Accidentes durante la jornada laboral:** 77,4%\n\n**Distribución por categoría de las enfermedades profesionales:**\n\n* **Enfermedades relacionadas con agentes biológicos:** 35,4%\n* **Enfermedades relacionadas con agentes químicos:** 22,7%\n* **Enfermedades relacionadas con agentes físicos:** 17,2%\n* **Enfermedades relacionadas con agentes ergonómicos:** 13,9%\n* **Otras enfermedades profesionales:** 10,8%\n\n**Total de accidentes de trabajo y enfermedades profesionales por categoría:**\n\n| Categoría | Accidentes de trabajo | Enfermedades profesionales | Total |\n|---|---|---|---|\n| Agentes biológicos | - | 5.790 | 5.790 |\n| Agentes químicos | - | 3.694 | 3.694 |\n| Agentes físicos | - | 2.800 | 2.800 |\n| Agentes ergonómicos | - | 2.261 | 2.261 |\n| Otras | - | 1.795 | 1.795 |\n| Accidentes in itinere | 140.533 | - | 140.533 |\n| Accidentes durante la jornada laboral | 481.689 | - | 481.689 |\n| **Total** | **622.222** | **16.340** | **638.562** |\n\n**Nota:** Los accidentes in itinere son aquellos que ocurren durante el trayecto entre el domicilio del trabajador y su lugar de trabajo, o viceversa."]
  },
])

```

```python
## Formular consulta ##
#consulta = "cuantos accidentes de trabajo hubo en chile en 2022, mostrar en forma de columnas separadas por comas,  las columnas deben ser algunas de las siguientes: AÑO,localidad o provincia,CIIU,Descripción,Personas trabajadoras con cobertura (promedio),Total de casos notificados,Casos con días de baja y secuelas incapacitantes,Personas trabajadoras fallecidas,Índice de incidencia ,Índice de incidencia de casos mortales"
#convo.send_message(consulta)
#salida = convo.last.text


salida =""
cantidad_de_anios = 2
paises =['Chile','Brasil']

for pais in paises:
  i=0
  while i<=cantidad_de_anios:
    consulta = "cuantos accidentes de trabajo hubo en "+ pais+"  en " + str(2012+i)+", mostrar en forma de columnas separadas por comas,  las columnas deben ser algunas de las siguientes: AÑO,localidad o provincia,CIIU,Descripción,Personas trabajadoras con cobertura (promedio),Total de casos notificados,Casos con días de baja y secuelas incapacitantes,Personas trabajadoras fallecidas,Índice de incidencia ,Índice de incidencia de casos mortales"
    convo.send_message(consulta)
    salida+= convo.last.text
    i+=1

```

```python
salida
```

```python
# Save the string to a CSV file
with open("data2.txt", "w", newline="") as f:
    f.write(salida)  # Or f.writelines(data_lines) if split into lines

print("Data saved to data.csv!")
```

## La aplicación Publicación de Resultados Electorales provee funcionalidades para que pueda ser utilizada por desarrolladores y desarrolladoras de software.
https://resultados.mininterior.gob.ar/desarrollo


```python
import requests
```

```python
url= 'https://resultados.mininterior.gob.ar/api/resultados/getResultados?anioEleccion=2019&tipoRecuento=1&tipoEleccion=1&categoriaId=2&distritoId=1&seccionProvincialId=0&seccionId=3&circuitoId=000039&mesaId=1244'
```

```python
headers = {
    "Authorization": "Bearer ",
    "Accept": "application/json"
}

response = requests.get(url, headers=headers)
```

```python
response
```

```python
data = response.json()
data
```

```python
import json
import pandas as pd

data_s = json.dumps(data['estadoRecuento'])
data_s = json.dumps(data['valoresTotalizadosPositivos'])
data_j = json.loads(data_s)
type(data_j)
```

```python
# Normalize and convert to DataFrame
df = pd.json_normalize(data_j)
df
```

```python
df.to_excel("data.xlsx")
```
